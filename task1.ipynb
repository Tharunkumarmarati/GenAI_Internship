## 1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"

Requirements:
Print "Login Successful" if both username and password match
Otherwise print "Invalid Credentials"

## 2. Pass / Fail Analyzer
Problem Statement:
Analyze student results.
marks = [45, 78, 90, 33, 60]

Requirements:
A student passes if marks ≥ 50
Count the total number of pass students
Count the total number of fail students
Print the final result clearly
Real-World Application: Academic evaluation systems


```python
marks = eval(input("Give input as list format"))
pass_count = 0
fail_count = 0
for i in marks:
    if i >= 50:
        pass_count += 1
    else:
        fail_count += 1
print(f'No.of students passed:{pass_count}\nNo.of students failed:{fail_count}')
```

names = [" Alice ", "bob", " CHARLIE "]

## 4. Message Length Analyzer
Problem Statement:
Analyze message sizes.
messages = ["Hi", "Welcome to the platform", "OK"]

Requirements:
Print the length of each message
Flag messages longer than 10 characters
Real-World Application: Text filtering and validation systems

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
    length = len(i)
    print(f'{i}:{length}')
    if length > 10:
        print(f"Flagged: '{i}' exceeds 10 characters")
```

## 5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries
Print the total error count
Real-World Application: Monitoring and log analysis systems

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0
for i in logs:
    if i == 'ERROR':
        error_count += 1
print("Total no.of errors:",error_count)
```
